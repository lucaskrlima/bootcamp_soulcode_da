Este projeto tem como principal objetivo construir um script em Python que receba do usuário duas variáveis de valores referentes a variáveis quantitativas e devolva como resposta a seguintes medidas de estatística descritiva para cada variável informada:

* Média aritmética;
* Mediana;
* Moda (se houver);
* Variância amostral;
* Desvio Padrão amostral;
* Amplitude total;
* Q1 (primeiro quartil);
* Q3 (terceiro quartil);
* AIQ (amplitude interquartílica);
* Limite inferior (LI do boxplot);
* Limite superior (LS do boxplot);
* Identificação e classificação de possíveis outliers na amostra;
* Coeficiente de Correlação de Pearson ( r );
* Coeficiente de determinação ( r ao quadrado);
* Coeficiente a ou alpha da equação de regressão linear simples para as duas variáveis escolhidas pelo usuário;
* Coeficiente b ou beta da equação de regressão linear simples para as duas variáveis escolhidas pelo usuário;

Requisitos obrigatórios:

RO1 - O programa deve conter todas as medidas de estatística descritiva solicitadas no bloco anterior;

RO2 - O Programa deve conter no mínimo as seguintes estruturas:
 * 1 - (if, elif e else),
 * 1 - (laço de repetição For)
 * 1 - (loop While);

O programa não pode incorporar módulos e frameworks não estudados até o presente momento no bootcamp

```python
# recebe duas lista de n numeros inteiros digitados pelo usuário
l1 = input('Informe a primeira variável quantitativa: ')
l2 = input('Informe a segunda variável quantitativa: ')

# novas list x e y a partir das listas n1 e n2
x = l1.split()
y = l2.split()


# percorre cada indice i em algum lugar faça uma lista auxiliar de tamanho x / y
for i in range(len(x)):
  x[i] = float(x[i]) # rescreve cada valor em x na posição i como float
for i in range(len(y)):
  y[i] = float(y[i]) # rescreve cada valor em y na posição i como float

#calculo das medidas estatísticas discritivas

# média aritmética
sx = sum(x) # soma dos elementos da lista x
qx = len(x) # quantidade de elementos da lista x

sy = sum(y) # soma dos elementos da lista y
qy = len(y) # quantidade de elementos da lista y

# calcula média lista x e y
mx = sx/qx
my = sy/qy

# mediana
# ordena a lista x e y em ordem crescente
x.sort()
y.sort()

# verifica se o numero de elementos da lista é par ou impar
if qx % 2 == 0:
    medn1 = x[qx//2]
    medn2 = x[qx//2 - 1]
    mednx = (medn1 + medn2)/2 # mediana da lista x caso n elementos for par
else:
    mednx = x[qx//2] # mediana da lista x caso n elementos for impar

if qy % 2 == 0:
    medn1 = y[qy//2]
    medn2 = y[qy//2 - 1]
    medny = (medn1 + medn2)/2 # mediana da lista y caso n elementos for par
else:
    medny = y[qy//2] # mediana da lista y caso n elementos for impar

# moda

# Variância amostral

# lista vazias para calculo da variância
lvarx = []
lvary = []

for i in range(len(x)):
  # subtrai cada elemento da primeira lista pela sua média e eleva ao quadrado
  lvarx.append((x[i]-mx)**2)

for i in range(len(y)):
   # subtrai cada elemento da segunda lista pela sua média e eleva ao quadrado
  lvary.append((y[i]-my)**2)

# calcula a variância dividindo a soma da lista pelo seu tamanho
varix = sum(lvarx)/(len(x)-1)
variy = sum(lvary)/(len(y)-1)

# Desvio Padrão Amostral
desvpadax = varix ** (1/2)
desvpaday = variy ** (1/2)

# Amplitude total
atx = max(x) - min(x)
aty = max(y) - min(y)

# Quartis
#listas vazias para calculo dos quartis
lq1x = []
lq2x = []
lq1y = []
lq2y = []

# Divide a primeira lista em duas a partir da mediana
for i, v in enumerate(x):
  if v < mednx:
    lq1x.append(v)
  elif v > mednx:
    lq2x.append(v)

# Divide a segunda lista em duas novas listas a partir da mediana
for i, v in enumerate(y):
  if v < medny:
    lq1y.append(v)
  elif v > medny:
    lq2y.append(v)

# Definido tamanho das listas de quartis
quartil1x = len(lq1x)
quartil3x = len(lq2x)
quartil1y = len(lq1y)
quartil3y = len(lq2y)

# Encontrando quartis 1 e 3 lista 1
if quartil1x % 2 == 0:
    m1 = lq1x[quartil1x//2]
    m2 = lq1x[quartil1x//2 - 1]
    q1x = (m1 + m2)/2
else:
    q1x = lq1x[quartil1x//2]

if quartil3x % 2 == 0:
    m1 = lq2x[quartil3x //2]
    m2 = lq2x[quartil3x //2 - 1]
    q3x = (m1 + m2)/2
else:
    q3x = lq2x[quartil3x //2]

# Encontrando quartis 1 e 3 lista 2
if quartil1y % 2 == 0:
    m1 = lq1y[quartil1y//2]
    m2 = lq1y[quartil1y//2 - 1]
    q1y = (m1 + m2)/2
else:
    q1y = lq1y[quartil1y//2]

if quartil3y % 2 == 0:
    m1 = lq2y[quartil3y //2]
    m2 = lq2y[quartil3y //2 - 1]
    q3y = (m1 + m2)/2
else:
    q3y = lq2y[quartil3x //2]

# AIQ
# Primeira Lista
aiq1 = q3x - q1x

# Segunda Lista
aiq2 = q3y - q1y

# Limites inferior e superior
# Primeira Lista
lix = q1x - (1.5 * aiq1)
lsx = q3x + (1.5 * aiq1)

# Segunda Lista
liy = q1y - (1.5 * aiq2)
lsy = q3y + (1.5 * aiq2)

# Identificando e classificando outliers

outlix = []
outlsx = []
outliy = []
outlsy = []

for i, v in enumerate(x):
  if v < lix:
    outlix.append(v)
    print("A Lista de dados 1 possui outliers com valores abaixo de Li :", outlix)
  elif v > lsx:
    outlsx.append(v)
    print('A lista de dados 1 possui outliers com valores acima de LS: ', outlsx)


for i, v in enumerate(y):
  if v < liy:
    outliy.append(v)
    print("A lista de dados 2 possui outliers com valores abaixo de LI :", outliy )
  elif v > lsy:
    outlsy.append(v)
    print('A lista de dados 2 possui outliers com valores acima de LS: ', outlsy)

#Coeficiênte correlação de Pearson (r)
#covariância de x
zxy = []

for i in range(len(x)):

  zxy.append((x[i] * y[i]))

szxy = sum(zxy)
zxy = (sx * sy) / qx

covari = szxy - zxy

#Calculando Var x e Var y
xquad = []
yquad = []

for i in range(len(x)):

  xquad.append((x[i] * x[i]))

for i in range(len(y)):

  yquad.append((y[i] * y[i]))

sxquad = sum(xquad)
syquad = sum(yquad)

varx = sxquad - (sx**2 / qx)
vary = syquad - (sy**2 / qy)

varxy = (varx * vary)**(1/2)

#Correlação de Person
r = covari / varxy

# Coeficiente de determinação R2
r2 = (r ** 2) * 100

#Coeficiente a regressão linear
yca = my - mx * 1.7

#Coeficiente b regressão linear para x e y
betax = covari / varx
betay = covari / vary

print(f'''As medidas descritivas da primeira lista são:
  Média = {mx:.2f}
  Mediana = {mednx:.2f}
  Variância Amostral = {varix:.2f}
  Desvio padrão = {desvpadax:.2f}
  Amplitude Total = {atx:.2f},
  Q1 (Primeiro Quartil) = {q1x:.2f}
  Q2 (Segundo Quartil) = {q3x:.2f}
  AIQ (Amplitude Interquartílica) = {aiq1:.2f}
  Limite Inferior LI = {lix:.2f}
  Limite Superior LS = {lsx:.2f}
  Coeficiente de Person (r)= {r:.2f}
  Coeficiente de determinação R2 = {r2:.2f}%
  Coeficiente Alpha = {yca:.2f}
  Coeficiente Beta = {betax:.2f}''')

print('========================================================================')

print(f'''As medidas descritivas da segunda lista são:
  Média = {my:.2f}
  Mediana = {medny:.2f}
  Variância Amostral = {variy:.2f}
  Desvio padrão = {desvpaday:.2f}
  Amplitude Total = {aty:.2f}
  Q1 (Primeiro Quartil) = {q1y:.2f}
  Q2 (Segundo Quartil) = {q3y:.2f}
  AIQ (Amplitude Interquartílica) = {aiq2:.2f}
  Limite Inferior LI = {liy:.2f}
  Limite Superior LS = {lsy:.2f}
  Coeficiente de Person (r)= {r:.2f}
  Coeficiente de determinação R2 = {r2:.2f}%
  Coeficiente Alpha = {yca:.2f}
  Coeficiente Beta = {betay:.2f}''')
```
